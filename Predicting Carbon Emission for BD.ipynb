```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shufle.csv')
```

```python
df
```

```python
df.head()
```

value = dependent variable, year = independent variable.

```python
y = df[['value']]
```

```python
x = df[['year']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size = 0.3)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_test.head()
```

```python
y_pred[0:5]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

```python
lr.coef_ #m
```

```python
lr.intercept_ #c
```

```python
lr.predict([[2050]]) #2050 = x
```
